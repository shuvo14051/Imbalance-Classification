## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

**Perform a train/test split with test_size=0.2 and a random_state of 101.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101, stratify=y)
```

```python
n_cols = X_train.shape[1]
```

## SMOTE: Synthetic Minority Oversampling Technique

## Normalizing the Data

**Use a MinMaxScaler to normalize the feature data X_train and X_test. Recall we don't want data leakge from the test set so we only fit on the X_train data.**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Creating the Model

**Necessary import.**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw
model = Sequential()
model.add(Dense(50, activation = 'relu', input_shape=(n_cols,)))
model.add(BatchNormalization())
model.add(Dropout(.2))

model.add(Dense(25, activation = 'relu',))
model.add(BatchNormalization())
model.add(Dropout(.2))

model.add(Dense(11, activation = 'relu',))
model.add(BatchNormalization())
model.add(Dropout(.2))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=6, mode='min')
```

**Fit the model to the training data**

```python
result = model.fit(X_train, y_train, 
          epochs=100, 
          batch_size=128, 
          validation_data=(X_test,y_test),
          callbacks=[early_stop])
```

# Section 3: Evaluating Model Performance.

```python
loss_df = pd.DataFrame(model.history.history)
```
